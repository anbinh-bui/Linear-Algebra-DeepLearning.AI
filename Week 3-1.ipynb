```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Scalar Multiplication & Sum of Vectors

```python
# Visualize vectors as arrows with label and color arguments

def plot_vectors(list_vector, list_label, list_color):
    _, ax = plt.subplots(figsize=(6, 6))
    ax.tick_params(axis='x', labelsize=5) # set axes' label and text size
    ax.tick_params(axis='x', labelsize=5)
    ax.set_xticks(np.arange(-10, 10)) # set axes' range size
    ax.set_yticks(np.arange(-10, 10))

    plt.axis([-10, 10, -10, 10]) # set axis plot range
    
    for i, v in enumerate(list_vector):
        sgn = 0.4 * np.array([[1] if i==0 else [i] for i in np.sign(v)])
        plt.quiver(v[0], v[1], color=list_color[i], angles='xy', scale_units='xy', scale=1) # @2D plot to show vector lines as arrows
        ax.text(v[0] - 0.2*sgn[0], v[1] - 0.2*sgn[1], list_label[i], fontsize=12, color=list_color[i])

    plt.grid()
    plt.gca().set_aspect("equal")
    plt.show()

v = np.array([[2], [3]])
plot_vectors([v], [f"$v$"], ["black"])
```

```python
# Scalar multiplication 
plot_vectors([v, 2*v, -2*v], [f"$v$", f"$2v$", f"$-2v$"], ["black", "green", "blue"]) 
```

```python
# Sum of vectors
u = np.array([[4], [2]])
plot_vectors([v, u, v + u], [f"$v$", f"$u$", f"$u+v$"], ["black", "green", "blue"])
```

## 2. Dot Product 

The dot product $x\cdot y$ is a scalar defined as:

$$x\cdot y = \sum_{i=1}^{n} x_iy_i = x_1y_1+x_2y_2+\ldots+x_ny_n \tag{1}$$

```python
x = np.array([1, 0, 2])
y = np.array([3, 4, 2])
x @ y
```

Matrix multiplication is the matrix version of dot product (vector version). 

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[3, 4], [5, 6]])
x @ y
```

Geometrically, the dot product is a Euclidean vector has both magnitude and direction. The dot product of two vectors $x$ and $y$ is defined by:

$$x\cdot y = \lvert x\rvert \lvert y\rvert \cos(\theta),\tag{2}$$

with $\theta$ as the angle between the two vectors. 
Dot product can be used to evaluate <b>vector similarity</b>. Similarity is defined as the cosine of the angle between them: similarity is maximum (cosine=1) when it's complete similar (same direction) and decreases as the angle increases. This is applied in NLP where we map words and phrases to a corresponding vectors of real numbers and compare them. 

Note: Orthogonality means 90 degree angle of difference, so two vectors are not at at all similar (cosine=0)

```python
i = np.array([5, 0, 0])
j = np.array([0, 5, 0])
plot_vectors([i, j],[f"$i$", f"$j$"], ["black", "blue"])
```
